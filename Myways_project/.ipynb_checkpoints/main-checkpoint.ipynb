```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pd.options.display.max_rows
```

```python
file = "Tourism.csv"
df = pd.read_csv(file)
df
```

```python
df_copy = df.copy()
df_copy = df_copy.sort_values(by=["2016 - Domestic"])
xd_states = df_copy["State/UT"][:-1]
y_domestic_2016 = df_copy["2016 - Domestic"][:-1]

df_copy = df_copy.sort_values(by=["2016 - Foreign"])
xf_states = df_copy["State/UT"][:-1]
y_foreign_2016 = df_copy["2016 - Foreign"][:-1]

font1 = {'family': 'serif', 'color': 'red', 'size': 15}
font2 = {'family': 'serif', 'color': 'blue', 'size': 10}

plt.subplot(2, 1, 1)
plt.bar(x=xd_states, height=y_domestic_2016, width=0.8, color="#82E0AA")
plt.xticks(rotation=90)
plt.title("Attraction: Top States 2016 (Domestic)", fontdict=font1, loc="center")
plt.xlabel("States", fontdict=font2)
plt.ylabel("no.of.tourists.visited", fontdict=font2)

plt.subplot(2, 1, 2)
plt.bar(x=xf_states, height=y_foreign_2016, width=0.8, color="#5DADE2")
plt.xticks(rotation=90)
plt.title("Attraction: Top States 2016 (Foreign)", fontdict=font1, loc="center")
plt.xlabel("States", fontdict=font2)
plt.ylabel("no.of.tourists.visited", fontdict=font2)

plt.subplots_adjust(hspace=0.9)
plt.show()
```
